```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
S0 = 1000
alpha = 0.1
T_max = 24
dt = 0.1
```

```python
# Tiempo
t = np.arange(0, T_max + dt, dt)
N = len(t)

# Inicializar arrays
S_euler = np.zeros(N)
S_analitico = S0 * np.exp(-alpha * t)

# Condición inicial
S_euler[0] = S0
```

```python
# Integración de Euler
for i in range(1, N):
    S_euler[i] = S_euler[i-1] - alpha * S_euler[i-1] * dt

# Gráfica
plt.plot(t, S_euler, label='Euler')
plt.plot(t, S_analitico, '--', label='Analítica')
plt.xlabel('Tiempo (meses)')
plt.ylabel('Población')
plt.title('Disminución de población')
plt.legend()
plt.grid()
plt.show()
```

- El método de Euler, con un paso pequeño (0.10.1 meses), es suficiente para modelar correctamente la dinámica de disminución.
- La población decrece de forma exponencial, como predice la teoría.
- Si hubieras usado un Δt mayor, la curva azul empezaría a separarse de la curva analítica, mostrando mayor error.
